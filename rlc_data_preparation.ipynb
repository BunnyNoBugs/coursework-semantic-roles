```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
rlc = pd.read_csv('annotated_texts_rlc.csv')
rlc = rlc[rlc['tags'].notna()]
rlc['corrected_text'] = rlc['corrected_text'].fillna('')
rlc['corrected_text'] = rlc['corrected_text'].apply(lambda x: x.strip())
rlc = rlc.sort_values(by=['original_text'])
rlc
```

```python
rlc.loc[59197]
```

```python
def get_tag_list(tags):
    subs = [
        ('/', ', '),
        ('\\', ', '),
        ('+', ', '),
        ('orpho', 'ortho'),
        ('orph', 'ortho')
    ]
    
    major_tags = {
        (
            'graph',
            'hyphen',
            'space',
            'ortho',
            'translit',
            'misspell'
        ): 'Ortho',
        (
            'deriv',
            'infl',
            'num',
            'gender',
            'altern',
            'morph'
        ): 'Morph',
        (
            'asp',
            'argstr',
            'passive',
            'tense',
            'mode',
            'refl',
            'agrnum',
            'argcase',
            'agrgender',
            'agrpers',
            'agrgerund',
            'gov',
            'ref',
            'conj',
            'wo',
            'neg',
            'aux',
            'brev',
            'com',
            'genneg',
            'impers',
            'syntax'
        ): 'Syntax',
        (
            'lex',
            'prep',
            'cs',
            'par',
            'idiom'
        ): 'Lex'
    }
    
    minor_tags = []
    for i in major_tags:
        minor_tags.extend(i)
        
    tags = tags.lower()
    for sub in subs:
        tags = tags.replace(sub[0], sub[1])
    tags = tags.split(', ')
    tags = [t if t in minor_tags else 'other' for t in tags]
    
    
    return tags
```

```python
rlc['tags'] = rlc['tags'].apply(get_tag_list)
rlc
```

```python
filtered_indices = [0]

curr_text = rlc.iloc[0]['original_text']
curr_idx = 0
for idx in range(len(rlc)):
    if rlc.iloc[idx]['corrected_text']:
        if rlc.iloc[idx]['original_text'] == curr_text:
            rlc.iloc[curr_idx]['tags'].extend(rlc.iloc[idx]['tags'])
        else:
            curr_text = rlc.iloc[idx]['original_text']
            curr_idx = idx
            filtered_indices.append(idx)
```

```python
rlc = rlc.iloc[filtered_indices]
rlc
```

```python
rlc = rlc.drop(['document_id', 'sentence_id', 'quote', 'correction', 'annotated', 'checked'], axis=1)
rlc['tags'] = rlc['tags'].apply(lambda x: list(set(x)))
rlc
```

```python
rlc_dev, rlc_test = train_test_split(rlc, test_size=5000, random_state=42)
rlc_dev = rlc_dev.drop(['tags'], axis=1)
rlc_test = rlc_test.drop(['corrected_text'], axis=1)
```

```python
rlc_dev_correct, rlc_dev_error = train_test_split(rlc_dev, test_size=0.5, random_state=42)
rlc_dev_correct['text'] = rlc_dev_correct['corrected_text']
rlc_dev_correct['label'] = 'correct'
rlc_dev_error['text'] = rlc_dev_error['original_text']
rlc_dev_error['label'] = 'incorrect'
rlc_dev = pd.concat([rlc_dev_correct, rlc_dev_error])
rlc_dev = rlc_dev.sample(frac=1, random_state=42)
rlc_dev = rlc_dev.drop(['original_text', 'corrected_text'], axis=1)
```

```python
rlc_dev.to_csv('rlc_dev.csv')
rlc_test.to_csv('rlc_test.csv')
rlc.to_csv('rlc_tags_united.csv')
```
